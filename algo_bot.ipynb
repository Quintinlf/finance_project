```python
# Enhanced Position and Strategy classes for Bayesian GP signals

import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import matplotlib.pyplot as plt

class BayesianPosition:
    """Enhanced Position class for Bayesian GP strategy"""
    def __init__(self, open_datetime, open_price, order_type, volume, sl, tp, 
                 signal_confidence=0.5, forecast_return=0.0, rsi_value=50):
        self.open_datetime = open_datetime
        self.open_price = open_price
        self.order_type = order_type
        self.volume = volume
        self.sl = sl
        self.tp = tp
        self.signal_confidence = signal_confidence
        self.forecast_return = forecast_return
        self.rsi_value = rsi_value
        self.close_datetime = None
        self.close_price = None
        self.profit = None
        self.status = 'open'
        self.close_reason = None

    def close_position(self, close_datetime, close_price, reason='manual'):
        self.close_datetime = close_datetime
        self.close_price = close_price
        self.close_reason = reason
        
        if self.order_type == 'buy':
            self.profit = (self.close_price - self.open_price) * self.volume
        else:
            self.profit = (self.open_price - self.close_price) * self.volume
        self.status = 'closed'

    def _asdict(self):
        return {
            'open_datetime': self.open_datetime,
            'open_price': self.open_price,
            'order_type': self.order_type,
            'volume': self.volume,
            'sl': self.sl,
            'tp': self.tp,
            'signal_confidence': self.signal_confidence,
            'forecast_return': self.forecast_return,
            'rsi_value': self.rsi_value,
            'close_datetime': self.close_datetime,
            'close_price': self.close_price,
            'profit': self.profit,
            'status': self.status,
            'close_reason': self.close_reason
        }

class BayesianStrategy:
    """Strategy class using Bayesian GP forecasts"""
    
    def __init__(self, df, starting_balance=10000, min_confidence=0.6):
        self.starting_balance = starting_balance
        self.positions = []
        self.data = df
        self.min_confidence = min_confidence
        self.current_balance = starting_balance
        
    def get_positions_df(self):
        if not self.positions:
            return pd.DataFrame()

        df = pd.DataFrame([position._asdict() for position in self.positions])
        
        if 'profit' not in df.columns:
            df['profit'] = 0.0
        else:
            df['profit'] = df['profit'].fillna(0.0)
            
        df['cumulative_pnl'] = df['profit'].cumsum() + self.starting_balance
        df['return_pct'] = df['profit'] / self.starting_balance * 100
        return df

    def add_position(self, position):
        self.positions.append(position)
        return True

    def close_tp_sl(self, data):
        """Close positions based on stop loss and take profit"""
        for pos in self.positions:
            if pos.status == 'open':
                current_price = data['Close']
                
                # Stop Loss logic
                if (pos.order_type == 'buy' and current_price <= pos.sl):
                    pos.close_position(data.name, pos.sl, 'stop_loss')
                elif (pos.order_type == 'sell' and current_price >= pos.sl):
                    pos.close_position(data.name, pos.sl, 'stop_loss')
                
                # Take Profit logic
                elif (pos.order_type == 'buy' and current_price >= pos.tp):
                    pos.close_position(data.name, pos.tp, 'take_profit')
                elif (pos.order_type == 'sell' and current_price <= pos.tp):
                    pos.close_position(data.name, pos.tp, 'take_profit')

    def has_open_positions(self):
        return any(pos.status == 'open' for pos in self.positions)

    def generate_bayesian_signal(self, data):
        """Generate trading signal using your Bayesian GP forecast"""
        try:
            # Get forecast for current ticker (you'd need to modify this for your specific data structure)
            ticker = getattr(data, 'ticker', 'UNKNOWN')
            
            # Simulate a Bayesian forecast (replace with actual forecast call)
            # In practice, you'd call: unified_bayesian_gp_forecast(ticker)
            
            # For now, use RSI + volatility as proxy
            rsi = data.get('rsi_14', 50)
            volatility = data.get('volatility', 0.02)
            
            # Simple Bayesian-like signal generation
            if rsi < 30:
                signal = 'BUY'
                confidence = 0.7 + (30 - rsi) / 100  # Higher confidence for lower RSI
                forecast_return = 0.02
            elif rsi > 70:
                signal = 'SELL'
                confidence = 0.7 + (rsi - 70) / 100
                forecast_return = -0.02
            else:
                signal = 'HOLD'
                confidence = 0.3
                forecast_return = 0.0
                
            return {
                'signal': signal,
                'confidence': min(confidence, 0.95),
                'forecast_return': forecast_return,
                'rsi': rsi
            }
            
        except Exception as e:
            return {
                'signal': 'HOLD',
                'confidence': 0.0,
                'forecast_return': 0.0,
                'rsi': 50
            }

    def logic(self, data):
        """Enhanced trading logic using Bayesian signals"""
        if not self.has_open_positions():
            
            # Generate Bayesian signal
            bayesian_signal = self.generate_bayesian_signal(data)
            
            # Only trade if confidence is above threshold
            if bayesian_signal['confidence'] >= self.min_confidence:
                
                current_price = data['Close']
                atr = data.get('atr_14', current_price * 0.02)  # Default 2% if ATR not available
                
                # Position sizing based on confidence (higher confidence = larger position)
                base_volume = 1000
                volume = base_volume * bayesian_signal['confidence']
                
                if bayesian_signal['signal'] == 'BUY':
                    sl = current_price - 2 * atr
                    tp = current_price + 3 * atr  # 3:2 risk-reward ratio
                    
                    position = BayesianPosition(
                        open_datetime=data.name,
                        open_price=current_price,
                        order_type='buy',
                        volume=volume,
                        sl=sl,
                        tp=tp,
                        signal_confidence=bayesian_signal['confidence'],
                        forecast_return=bayesian_signal['forecast_return'],
                        rsi_value=bayesian_signal['rsi']
                    )
                    self.add_position(position)
                    
                elif bayesian_signal['signal'] == 'SELL':
                    sl = current_price + 2 * atr
                    tp = current_price - 3 * atr
                    
                    position = BayesianPosition(
                        open_datetime=data.name,
                        open_price=current_price,
                        order_type='sell',
                        volume=volume,
                        sl=sl,
                        tp=tp,
                        signal_confidence=bayesian_signal['confidence'],
                        forecast_return=bayesian_signal['forecast_return'],
                        rsi_value=bayesian_signal['rsi']
                    )
                    self.add_position(position)

    def run_backtest(self):
        """Run the complete backtest"""
        print(f"üöÄ Starting Bayesian GP Backtest")
        print(f"üìä Initial Balance: ${self.starting_balance:,.2f}")
        print(f"üéØ Min Confidence Threshold: {self.min_confidence:.1%}")
        print("="*50)
        
        for i, (timestamp, data) in enumerate(self.data.iterrows()):
            self.close_tp_sl(data)
            self.logic(data)
            
            # Progress update every 50 days
            if i % 50 == 0:
                open_positions = sum(1 for p in self.positions if p.status == 'open')
                total_profit = sum(p.profit for p in self.positions if p.profit is not None)
                print(f"Day {i}: Open positions: {open_positions}, P&L: ${total_profit:+.2f}")
        
        # Close any remaining open positions at the end
        if self.positions:
            final_data = self.data.iloc[-1]
            for pos in self.positions:
                if pos.status == 'open':
                    pos.close_position(final_data.name, final_data['Close'], 'backtest_end')
        
        return self.get_positions_df()

    def analyze_performance(self, positions_df):
        """Analyze backtest performance"""
        if positions_df.empty:
            print("‚ùå No trades executed during backtest period")
            return {}
        
        # Performance metrics
        total_trades = len(positions_df)
        winning_trades = len(positions_df[positions_df['profit'] > 0])
        losing_trades = len(positions_df[positions_df['profit'] < 0])
        
        win_rate = winning_trades / total_trades if total_trades > 0 else 0
        total_pnl = positions_df['profit'].sum()
        final_balance = self.starting_balance + total_pnl
        
        avg_win = positions_df[positions_df['profit'] > 0]['profit'].mean() if winning_trades > 0 else 0
        avg_loss = positions_df[positions_df['profit'] < 0]['profit'].mean() if losing_trades > 0 else 0
        
        # Risk metrics
        max_drawdown = (positions_df['cumulative_pnl'].cummax() - positions_df['cumulative_pnl']).max()
        sharpe_ratio = positions_df['return_pct'].mean() / positions_df['return_pct'].std() if positions_df['return_pct'].std() > 0 else 0
        
        performance = {
            'total_trades': total_trades,
            'winning_trades': winning_trades,
            'losing_trades': losing_trades,
            'win_rate': win_rate,
            'total_pnl': total_pnl,
            'final_balance': final_balance,
            'return_pct': (final_balance - self.starting_balance) / self.starting_balance * 100,
            'avg_win': avg_win,
            'avg_loss': avg_loss,
            'profit_factor': abs(avg_win * winning_trades / avg_loss / losing_trades) if avg_loss != 0 and losing_trades > 0 else float('inf'),
            'max_drawdown': max_drawdown,
            'sharpe_ratio': sharpe_ratio
        }
        
        # Print performance report
        print("\nüéØ BAYESIAN GP BACKTEST RESULTS")
        print("="*50)
        print(f"üìà Total Trades: {total_trades}")
        print(f"üü¢ Winning Trades: {winning_trades} ({win_rate:.1%})")
        print(f"üî¥ Losing Trades: {losing_trades}")
        print(f"üí∞ Total P&L: ${total_pnl:+,.2f}")
        print(f"üè¶ Final Balance: ${final_balance:,.2f}")
        print(f"üìä Total Return: {performance['return_pct']:+.2f}%")
        print(f"‚ö° Avg Win: ${avg_win:+.2f}")
        print(f"‚ö° Avg Loss: ${avg_loss:+.2f}")
        print(f"üéØ Profit Factor: {performance['profit_factor']:.2f}")
        print(f"üìâ Max Drawdown: ${max_drawdown:.2f}")
        print(f"üìè Sharpe Ratio: {sharpe_ratio:.2f}")
        
        return performance

def plot_backtest_results(positions_df, performance):
    """Plot backtest results"""
    if positions_df.empty:
        print("No data to plot")
        return
    
    fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 10))
    
    # Equity curve
    ax1.plot(positions_df['close_datetime'], positions_df['cumulative_pnl'], 'b-', linewidth=2)
    ax1.set_title('Equity Curve')
    ax1.set_ylabel('Portfolio Value ($)')
    ax1.grid(True, alpha=0.3)
    
    # Profit distribution
    ax2.hist(positions_df['profit'], bins=20, alpha=0.7, color='green', edgecolor='black')
    ax2.set_title('Profit Distribution')
    ax2.set_xlabel('Profit per Trade ($)')
    ax2.set_ylabel('Frequency')
    ax2.grid(True, alpha=0.3)
    
    # Confidence vs Profit scatter
    ax3.scatter(positions_df['signal_confidence'], positions_df['profit'], 
               c=positions_df['profit'], cmap='RdYlGn', alpha=0.6)
    ax3.set_title('Signal Confidence vs Profit')
    ax3.set_xlabel('Signal Confidence')
    ax3.set_ylabel('Profit ($)')
    ax3.grid(True, alpha=0.3)
    
    # Monthly returns
    positions_df['month'] = pd.to_datetime(positions_df['close_datetime']).dt.to_period('M')
    monthly_returns = positions_df.groupby('month')['profit'].sum()
    ax4.bar(range(len(monthly_returns)), monthly_returns.values, 
           color=['green' if x > 0 else 'red' for x in monthly_returns.values])
    ax4.set_title('Monthly P&L')
    ax4.set_ylabel('Monthly Profit ($)')
    ax4.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

print("‚úÖ Enhanced Bayesian backtesting classes loaded!")
```

```python
# Add this to your notebook after the Bollinger Bands cell
def backtest_portfolio_strategies(portfolio, period="2y", min_confidence=0.6):
    """
    Backtest Bayesian GP strategies for entire portfolio
    """
    print("üéØ RUNNING PORTFOLIO BACKTESTS")
    print("="*60)
    
    portfolio_backtest_results = {}
    
    for ticker, shares in portfolio.items():
        print(f"\nüöÄ Backtesting {ticker.upper()}...")
        
        try:
            # Get historical data for backtesting
            df = yf.download(ticker, period=period, interval="1d", progress=False)
            
            if df.empty or len(df) < 100:
                print(f"‚ùå Insufficient data for {ticker}")
                continue
            
            # Calculate technical indicators for backtesting
            df['return'] = df['Close'].pct_change()
            
            # RSI calculation
            delta = df['Close'].diff()
            gain = delta.where(delta > 0, 0).rolling(window=14).mean()
            loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()
            rs = gain / loss
            df['rsi_14'] = 100 - (100 / (1 + rs))
            
            # ATR calculation (simplified)
            high_low = df['High'] - df['Low']
            high_close = np.abs(df['High'] - df['Close'].shift())
            low_close = np.abs(df['Low'] - df['Close'].shift())
            tr = np.maximum(high_low, np.maximum(high_close, low_close))
            df['atr_14'] = tr.rolling(window=14).mean()
            
            # Volatility
            df['volatility'] = df['return'].rolling(window=20).std()
            
            df = df.dropna()
            
            if len(df) < 50:
                print(f"‚ö†Ô∏è Not enough clean data for {ticker}")
                continue
            
            # Run backtest
            strategy = BayesianStrategy(df, starting_balance=10000, min_confidence=min_confidence)
            positions_df = strategy.run_backtest()
            
            if not positions_df.empty:
                performance = strategy.analyze_performance(positions_df)
                portfolio_backtest_results[ticker] = {
                    'positions': positions_df,
                    'performance': performance,
                    'strategy': strategy
                }
                print(f"‚úÖ {ticker.upper()} backtest complete!")
            else:
                print(f"‚ö†Ô∏è No trades generated for {ticker}")
                
        except Exception as e:
            print(f"‚ùå Error backtesting {ticker}: {str(e)}")
    
    # Portfolio summary
    print(f"\nüèÜ PORTFOLIO BACKTEST SUMMARY")
    print("="*80)
    
    if portfolio_backtest_results:
        total_return = 0
        total_trades = 0
        winning_strategies = 0
        
        for ticker, results in portfolio_backtest_results.items():
            perf = results['performance']
            total_return += perf['return_pct']
            total_trades += perf['total_trades']
            if perf['return_pct'] > 0:
                winning_strategies += 1
                
            print(f"{ticker:6} | Return: {perf['return_pct']:+6.1f}% | Trades: {perf['total_trades']:3d} | Win Rate: {perf['win_rate']:5.1%}")
        
        avg_return = total_return / len(portfolio_backtest_results)
        strategy_win_rate = winning_strategies / len(portfolio_backtest_results)
        
        print("-"*80)
        print(f"üìä Portfolio Avg Return: {avg_return:+.1f}%")
        print(f"üéØ Total Trades Executed: {total_trades}")
        print(f"üèÜ Winning Strategies: {winning_strategies}/{len(portfolio_backtest_results)} ({strategy_win_rate:.1%})")
    
    return portfolio_backtest_results

# Run the portfolio backtest
backtest_results = backtest_portfolio_strategies(portfolio, period="1y", min_confidence=0.65)
```

```python

import yfinance as yf
import numpy as np
import pandas as pd
import sys
import os
sys.path.append(r'c:\\Users\\Windows User\\My_folder\\Finance_project')

try:
    from trading_functions import unified_bayesian_gp_forecast
    print("‚úÖ Successfully imported unified_bayesian_gp_forecast")
except ModuleNotFoundError as e:
    print(f"‚ùå ModuleNotFoundError: {e}. Please ensure trading_functions.py is in the correct folder.")

portfolio = {
    'czoof': 0.2,
    'FCNVX': 0.995,
    'fdig': 0.925,
    'fdkvx': 0.862,
    'feddx': 0.961,
    'ffrhx': 3.247,
    'fifnx': 0.969,
    'fitlx': 1.151,
    'flowx': 0.838,
    'fsenx': 0.326,
    'LION': 2,
    'OLLI': 0.175,
    'UNP': 0.1,
    'VZ': 0.5,
    'WPP': 0.3
}
ticker_list = list(portfolio.keys())

# --- Adapted Bollinger Bands Strategy for Portfolio ---
class Position:
    def __init__(self, open_datetime, open_price, order_type, volume, sl, tp):
        self.open_datetime = open_datetime
        self.open_price = open_price
        self.order_type = order_type
        self.volume = volume
        self.sl = sl
        self.tp = tp
        self.close_datetime = None
        self.close_price = None
        self.profit = None
        self.status = 'open'

    def close_position(self, close_datetime, close_price):
        self.close_datetime = close_datetime
        self.close_price = close_price
        self.profit = (self.close_price - self.open_price) * self.volume if self.order_type == 'buy' \
            else (self.open_price - self.close_price) * self.volume
        self.status = 'closed'

    def _asdict(self):
        return {
            'open_datetime': self.open_datetime,
            'open_price': self.open_price,
            'order_type': self.order_type,
            'volume': self.volume,
            'sl': self.sl,
            'tp': self.tp,
            'close_datetime': self.close_datetime,
            'close_price': self.close_price,
            'profit': self.profit,
            'status': self.status,
        }

class Strategy:
    def __init__(self, df, starting_balance, volume):
        self.starting_balance = starting_balance
        self.volume = volume
        self.positions = []
        self.data = df

    def get_positions_df(self):
        df = pd.DataFrame([position._asdict() for position in self.positions])
        df['pnl'] = df['profit'].cumsum() + self.starting_balance
        return df

    def add_position(self, position):
        self.positions.append(position)

    def trading_allowed(self):
        for pos in self.positions:
            if pos.status == 'open':
                return False
        return True

    def run(self):
        for i, data in self.data.iterrows():
            # Optionally, use Bayesian GP forecast as a filter
            # forecast = unified_bayesian_gp_forecast(ticker) # Uncomment if you want to use it

            # Only open a position if allowed and signal is 'sell' or 'buy'
            if data[('signal', '')] == 'sell' and self.trading_allowed():
                sl = data[('Close', ticker.upper())] + 3 * data[('sd', '')]
                tp = data[('Close', ticker.upper())] - 2 * data[('sd', '')]
                self.add_position(Position(data[('time', '')], data[('Close', ticker.upper())], 'sell', self.volume, sl, tp))
            elif data[('signal', '')] == 'buy' and self.trading_allowed():
                sl = data[('Close', ticker.upper())] - 3 * data[('sd', '')]
                tp = data[('Close', ticker.upper())] + 2 * data[('sd', '')]
                self.add_position(Position(data[('time', '')], data[('Close', ticker.upper())], 'buy', self.volume, sl, tp))

            # Manage open positions
            for pos in self.positions:
                if pos.status == 'open':
                    price = data[('Close', ticker.upper())]
                    if (pos.sl >= price and pos.order_type == 'buy'):
                        pos.close_position(data[('time', '')], pos.sl)
                    elif (pos.sl <= price and pos.order_type == 'sell'):
                        pos.close_position(data[('time', '')], pos.sl)
                    elif (pos.tp <= price and pos.order_type == 'buy'):
                        pos.close_position(data[('time', '')], pos.tp)
                    elif (pos.tp >= price and pos.order_type == 'sell'):
                        pos.close_position(data[('time', '')], pos.tp)
        return self.get_positions_df()

# --- Run for each ticker in your portfolio ---
results = {}
for ticker in ticker_list:
    print(f"\nüîé Processing {ticker} ...")
    df = yf.download(ticker, period="1y", interval="1d", progress=False)
    if df.empty or len(df) < 50:
        print(f"‚ùå Not enough data for {ticker}")
        continue

    # Add calculated columns
    df[('time', '')] = df.index
    df[('sma', '')] = df[('Close', ticker.upper())].rolling(20).mean()
    df[('sd', '')] = df[('Close', ticker.upper())].rolling(20).std()
    df[('lb', '')] = df[('sma', '')] - 2 * df[('sd', '')]
    df[('ub', '')] = df[('sma', '')] + 2 * df[('sd', '')]
    df.dropna(inplace=True)

    # Ensure columns are MultiIndex
    # Signal assignment using correct keys
    def find_signal(close, lb, ub):
        if close < lb:
            return 'buy'
        elif close > ub:
            return 'sell'
        else:
            return 'hold'

    df[('signal', '')] = df.apply(
        lambda row: find_signal(row[('Close', ticker.upper())], row[('lb', '')], row[('ub', '')]),
        axis=1
    )

    strat = Strategy(df, 10000, 1000)
    result = strat.run()
    results[ticker] = result

    print(f"‚úÖ {ticker} backtest complete! Trades: {len(result)}")

def plot_forecast_and_backtest(portfolio_forecasts, results):
    buy_signals = 0
    sell_signals = 0
    hold_signals = 0

    for ticker_sym, forecast_data in portfolio_forecasts.items():
        print(f"\nüìà Generating plots for {ticker_sym.upper()}...")

        if 'error' in forecast_data:
            print(f"‚ö†Ô∏è Skipping {ticker_sym.upper()} due to error: {forecast_data['error']}")
            continue

        result = results.get(ticker_sym)
        if result is None or not isinstance(result, pd.DataFrame):
            print(f"‚ö†Ô∏è No backtest result for {ticker_sym.upper()}, skipping plot.")
            continue

        try:
            df = yf.download(ticker_sym, period="1y", interval="1d", progress=False)
            if df.empty or len(df) < 50:
                print(f"‚ö†Ô∏è Not enough price data for {ticker_sym.upper()}, skipping plot.")
                continue

            df = df.copy()
            df['time'] = df.index
            df['sma'] = df['Close'].rolling(20).mean()
            df['sd'] = df['Close'].rolling(20).std()
            df['lb'] = df['sma'] - 2 * df['sd']
            df['ub'] = df['sma'] + 2 * df['sd']
            df.dropna(inplace=True)

            def find_signal(row):
                close = float(row['Close'])
                lb = float(row['lb'])
                ub = float(row['ub'])
                if close < lb:
                    return 'buy'
                elif close > ub:
                    return 'sell'
                else:
                    return 'hold'

            df['signal'] = df.apply(find_signal, axis=1)

        except Exception as e:
            print(f"‚ö†Ô∏è Error downloading or processing data for {ticker_sym.upper()}: {e}")
            continue

        # --- Price & Bollinger Bands Plot ---
        fig, ax = plt.subplots(figsize=(14, 6))
        ax.plot(df['time'], df['Close'], label='Close')
        ax.plot(df['time'], df['sma'], label='SMA')
        ax.plot(df['time'], df['lb'], label='Lower Band')
        ax.plot(df['time'], df['ub'], label='Upper Band')

        buy_mask = df['signal'] == 'buy'
        sell_mask = df['signal'] == 'sell'
        ax.scatter(df['time'][buy_mask], df['Close'][buy_mask], label='Buy', color='green', marker='^')
        ax.scatter(df['time'][sell_mask], df['Close'][sell_mask], label='Sell', color='red', marker='v')

        ax.set_title(f"{ticker_sym.upper()} Price & Bollinger Bands")
        ax.set_xlabel('Date')
        ax.set_ylabel('Price')
        ax.legend()
        plt.show()

        # --- PnL Plot ---
        if 'pnl' in result.columns:
            plt.figure(figsize=(10, 4))
            plt.plot(result['close_datetime'], result['pnl'], label='PnL')
            plt.title(f"{ticker_sym.upper()} Backtest PnL")
            plt.xlabel('Close Date')
            plt.ylabel('PnL')
            plt.legend()
            plt.show()

    return
```
